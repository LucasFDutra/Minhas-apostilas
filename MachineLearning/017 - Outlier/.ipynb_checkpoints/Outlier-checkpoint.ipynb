# **OUTLIERS**

- Valores anormais.
- Valores fora do padrão (afastados da média)
- Causas:
    - Acaso
    - Erro no preenchimento dos dados.
    - Fraudes.
- Tratamento
    - Remover o registro
    - Não fazer nada (o que atrapalha principalmente na regressão)
    - Substituir o valor.

## DETECÇÃO DE OUTLIERS COM GRÁFICO DE DISPERSÃO

Os gráficos de dispersão são feito com pares de atributos.

---
- Link para aprender mais:
    - [Outliers em diagramas de dispersão - Khan Academy](https://pt.khanacademy.org/math/cc-eighth-grade-math/cc-8th-data/cc-8th-interpreting-scatter-plots/a/outliers-in-scatter-plots)
---
- **Base de Dados**:
    - credit_data.csv
    
Primeiramente vamos fazer um tratamento na base de dados.

```python
#----------------Coloca a base de dados na variável 'base'--------------------#
base = pd.read_csv('../input/credit_data.csv')

#------------------------------Trata os dados---------------------------------#
# nesse caso foram retirados registros com nan                                #
# e substituindo valores negativos de idade pela média                        #
#-----------------------------------------------------------------------------#
base = base.dropna()
base.loc[base.age < 0, 'age'] = 40.92
```

Agora vamos exibir a dispersão para os pares income x age

```python
#---------------------------income x age--------------------------------------#
plt.scatter(base.iloc[:,1], base.iloc[:,2])
```

Veja que não houve outliers entre as variáveis income e age.

Agora vamos exibir a dispersão para os pares income x loan

```python
#---------------------------income x loan-------------------------------------#
plt.scatter(base.iloc[:,1], base.iloc[:,3])
```

Veja que não houve outlier entre as variáveis income e loan

Agora vamos exibir a dispersão para os pares age x loan

```python
#----------------------------age x loan---------------------------------------#
plt.scatter(base.iloc[:,2], base.iloc[:,3])
```

Também não houveram outliers entre age e loan.
Se não tivéssemos feito a substituição dos valores de idade negativas pela média, existiriam outliers, mas o restante dos dados não destoam muito.

Veja como ficariam os gráficos sem efetuarmos a troca:

```python
#----------------Coloca a base de dados na variável 'base'--------------------#
base = pd.read_csv('../input/credit_data.csv')

#------------------------------Trata os dados---------------------------------#
# nesse caso foram retirados registros com nan                                #
#-----------------------------------------------------------------------------#
base = base.dropna()
#---------------------------income x age--------------------------------------#
plt.scatter(base.iloc[:,1], base.iloc[:,2])
```

```python
#---------------------------income x loan-------------------------------------#
plt.scatter(base.iloc[:,1], base.iloc[:,3])
```

```python
#----------------------------age x loan---------------------------------------#
plt.scatter(base.iloc[:,2], base.iloc[:,3])
```

## DETECÇÃO DE OUTLIERS COM PYOD

Os dois métodos apresentados anteriormente são de certa forma "manuais", a PyOD é uma biblioteca especifica para detecção de outliers. Vendo a [documentação](https://pyod.readthedocs.io/en/latest/) da biblioteca é possível verificar que existem diversos métodos para a detecção de outliers, nesse exemplo vou utilizar o método KNN.

- **Base de Dados**:
    - credit_data.csv
 Primeiramente instalamos o pyod.

```python
!pip install pyod
```

Primeiramente passamos a base de dados.

```python
#----------------Coloca a base de dados na variável 'base'--------------------#
base = pd.read_csv('../input/credit_data.csv')
base = base.dropna()
```

Criamos um detector, mais especificamente ele irá se adequar para identificar outliers entre a relação dos atributos 1, 2 e 3.

> OBS.: Veja que na boxplot podíamos detectar outliers somente entre uma variável, no gráfico de dispersão pode-se detectar entre duas, com PyOD podemos passar para detectar outliers entre mais de dois atributos

```python
#----------Criando um detector adequando para os atributo de 1 ao 3-----------#
from pyod.models.knn import KNN
detector = KNN()
detector.fit(base.iloc[:,1:4])

#---------------------------Encontrando os outliers---------------------------#
# quando previsões = 0: não é outlier                                         #
# quando previsões = 1: é outlier                                             #
#-----------------------------------------------------------------------------#
previsoes = detector.labels_

#-------------------Verificando a confiança nessa detecção--------------------#
# quanto maior o valor maior a chance de ser um outlier                       #
#-----------------------------------------------------------------------------#
confianca_previsoes = detector.decision_scores_
```

Identificando e armazenando os registros que são considerados outliers.

```python
#-----------------Identificando os indices que são outlier--------------------#
outliers = []
for i in range(len(previsoes)):
    #print(previsoes[i])
    if previsoes[i] == 1:
        outliers.append(i)

#----Capturando as linahs correspondentes a outliers na base e armazenando----#
lista_outliers = base.iloc[outliers, :]
print(lista_outliers[:10])
```

Agora em posse da informação de quais registros são outliers pode-se tratar melhor os dados.
